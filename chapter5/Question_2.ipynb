```python
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
```

```python
def calc_prob_in_sample(n: int) -> float:
    return 1 - (1 - 1/n) ** n 
```

```python

n_list = range (100, 100_001)
probs = [calc_prob_in_sample(n) for n in n_list]
```

```python
sns.lineplot(x=n_list, y=probs)
plt.show()
```

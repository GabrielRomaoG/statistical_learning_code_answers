### d)

```python
def Power3(base: int, exponent: int) -> int:
    result = base ** exponent

    return result
```

### e)

```python
X = range(1, 11)
y = [Power3(item,exponent=2) for item in X]

plt.scatter(
    x=X,
    y=y
)
```

### f)

```python
def PlotPower(int_list, exponent):
    y = [Power3(item,exponent=2) for item in int_list]
    plt.scatter(
        x=int_list,
        y=y
    )
PlotPower(np.arange(1, 11), 3)
```

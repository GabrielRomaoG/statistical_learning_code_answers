## Auto Dataset

Gas mileage, horsepower, and other information for 392 vehicles.

A data frame with 392 observations on the following 9 variables.

- mpg: miles per gallon
- cylinders: Number of cylinders between 4 and 8
- displacement: Engine displacement (cu. inches)
- horsepower: Engine horsepower
- weight: Vehicle weight (lbs.)
- acceleration: Time to accelerate from 0 to 60 mph (sec.)
- year: Model year (modulo 100)
- origin: Origin of car (1. American, 2. European, 3. Japanese)
- name: Vehicle name

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os


#path of the datasets
os.chdir("../Statistical Learning Answers/..")
dataset_path = os.path.join(os.getcwd(),"DataSets")
```

```python
auto_df  = pd.read_csv(f'{dataset_path}/auto.csv')
auto_df
```

```python
describe_df = auto_df.describe().transpose()
describe_df
```

```python
auto_df.info()
```

```python
# horsepower is registered as object type

# testing if there are any values that aren't numbers
auto_df.loc[~auto_df.horsepower.apply(lambda x: x.isnumeric())]
```

```python
# For now i will change "?" for None

auto_df['horsepower'] = pd.to_numeric(auto_df['horsepower'].replace('?', None))
auto_df.loc[auto_df.horsepower.isna()]
```

```python
# Quantitave predictors
auto_df.select_dtypes('number').columns
```

```python
# Qualitative predictors
auto_df.select_dtypes('object').columns
```

```python
# Removing the 10th through 85th observation

auto_df_filtered1 = auto_df.drop(auto_df.index[10:85]).copy()

# Statistics without these observations 
describe_df_filt = auto_df_filtered1.describe().transpose()
describe_df_filt['range'] = describe_df_filt['max'] - describe_df_filt['min']

print('Statistics of the filterd dataframe')
print(describe_df_filt[['range', 'mean', 'std']])

#Comparing with the statistics of the original dataframe

print('Statistics of the original Dataframe')
describe_df['range'] = describe_df['max'] - describe_df['min']
print(describe_df[['range', 'mean', 'std']])
```

```python
# Distribution of MPG column
plt.figure(figsize=(12, 6))
sns.histplot(auto_df['mpg'], bins=20, kde=True)
plt.title('Distribution of MPG')
plt.xlabel('MPG')
plt.ylabel('Frequency')
plt.show()
```

```python
# Relationship between Weight and MPG
plt.figure(figsize=(10, 6))
sns.scatterplot(x='weight', y='mpg', data=auto_df)
plt.title('Scatter Plot of Weight vs. MPG')
plt.xlabel('Weight')
plt.ylabel('MPG')
plt.show()
```

```python
# Correlation between all the numerical columns
plt.figure(figsize=(10, 6))
sns.heatmap(auto_df.iloc[:, :-1].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

As we can see, cylinders, displacement, horsepower and weight have negative correlation with mpg, that means increasing these variables will decrease mpg.
